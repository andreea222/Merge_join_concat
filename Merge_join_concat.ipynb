```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data1 = {'Name':['Andrei', 'Marius', 'Alex', 'Stefan'], 
        'Age':[27, 24, 22, 32], 
        'Address':['Bucuresti', 'Bucuresti', 'Brasov', 'Brasov'], 
        'Qualification':['Msc', 'MA', 'MCA', 'Phd'], 'Mobile No': [97, 91, 58, 76]} 
```

```python
data2 = {'Name':['Andrei', 'Marius', 'Robert', 'Radu'], 
        'Age':[27, 24, 22, 32], 
        'Address':['Bucuresti', 'Bucuresti', 'Brasov', 'Brasov'], 
        'Qualification':['Msc', 'MA', 'MCA', 'Phd'], 'Salary':[1000, 2000, 3000, 4000]} 
```

```python
df1 = pd.DataFrame(data1)
df1
```

```python
df2 = pd.DataFrame(data2)
df2
```

```python
frames = [df1, df2]
frames
```

```python
res = pd.concat(frames , ignore_index=True, sort=False)
res
# default: join='outer' : include all the records
```

```python
# Now we set axes join = inner for intersection of dataframe
# and use axis = 1
res2 = pd.concat([df1, df2], sort=False)
res2
```

```python
res3=df1.append(df2, sort=False)
res3
```

```python
# we can concat a df with a series which we give a name to be used as column name
s1 = pd.Series([1000, 2000, 3000, 4000], name='Salary')
s1
```

```python
res4 = pd.concat([df1, s1], axis=1)
res4
```

```python
res5 = df1.join(df2, how= 'inner' )
```
